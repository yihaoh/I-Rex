```python
import json
# import matplotlib as plt
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

def draw_three_line_subplots(list1, list2, list3, list4, list5, list6):
    """
    Draw a plot with 3 separate subplots. Each subplot will display two lists of integers
    represented by lines.
    
    Parameters:
    list1, list2 (list of int): First pair of lists to plot in the first subplot.
    list3, list4 (list of int): Second pair of lists to plot in the second subplot.
    list5, list6 (list of int): Third pair of lists to plot in the third subplot.
    """
    fig, axes = plt.subplots(3, 1, figsize=(10, 12))  # Create 3 subplots

    # Plot first pair
    axes[0].plot(list(range(len(list1))), list1, marker='o', linestyle='-', color='b', label='List 1')
    axes[0].plot(list(range(len(list2))), list2, marker='x', linestyle='-', color='r', label='List 2')
    axes[0].set_title('Pair 1')
    axes[0].set_xlabel('Index')
    axes[0].set_ylabel('Value')
    axes[0].legend()
    axes[0].grid(True)

    # Plot second pair
    axes[1].plot(list(range(len(list3))), list3, marker='o', linestyle='-', color='b', label='List 3')
    axes[1].plot(list(range(len(list4))), list4, marker='x', linestyle='-', color='r', label='List 4')
    axes[1].set_title('Pair 2')
    axes[1].set_xlabel('Index')
    axes[1].set_ylabel('Value')
    axes[1].legend()
    axes[1].grid(True)

    # Plot third pair
    axes[2].plot(list(range(len(list5))), list5, marker='o', linestyle='-', color='b', label='List 5')
    axes[2].plot(list(range(len(list6))), list6, marker='x', linestyle='-', color='r', label='List 6')
    axes[2].set_title('Pair 3')
    axes[2].set_xlabel('Index')
    axes[2].set_ylabel('Value')
    axes[2].legend()
    axes[2].grid(True)

    # Adjust layout
    plt.tight_layout()
    plt.show()
```

## Bloom filter charts

```python
# 1G data, vary page size

files = ['results_special/s1_p20/q2-bloom-result.json', 'results_special/s1_p50/q2-bloom-result.json', 'results_special/s1_p100/q2-bloom-result.json']
lines = []
for file in files:
    with open(file) as f:
        d = json.load(f)
    lines.append(d['q2_join_filter']['page_query_time'])
    lines.append(d['q2_join_filter']['naive_page_query_time'])

fig, axes = plt.subplots(3, 1, figsize=(5, 6))  # Create 3 subplots

# Plot first pair
axes[0].plot(list(range(len(lines[0]))), lines[0], marker='o', linestyle='-', color='b', label=f'Bloom: [{int(min(lines[0]))}, {int(max(lines[0]))}]')
axes[0].plot(list(range(len(lines[1]))), lines[1], marker='x', linestyle='-', color='r', label=f'Base: [{int(min(lines[1]))}, {int(max(lines[1]))}]')
# axes[0].set_title('Page Fetch Time (db: 1GB, page: 20)', fontsize=18)
axes[0].set_title('Page Fetch Time (page: 20, 50, 100)', fontsize=18)
# axes[0].set_xlabel('Page', fontsize=18)
axes[0].set_ylabel('Time (ms)', fontsize=18)
axes[0].tick_params(labelsize=15)
axes[0].legend(fontsize=13)
axes[0].grid(True)

# Plot second pair
axes[1].plot(list(range(len(lines[2]))), lines[2], marker='o', linestyle='-', color='b', label=f'Bloom: [{int(min(lines[2]))}, {int(max(lines[2]))}]')
axes[1].plot(list(range(len(lines[3]))), lines[3], marker='x', linestyle='-', color='r', label=f'Base: [{int(min(lines[3]))}, {int(max(lines[3]))}]')
# axes[1].set_title('Page Fetch Time (db: 1GB, page: 50)', fontsize=18)
# axes[1].set_xlabel('Page', fontsize=18)
axes[1].set_ylabel('Time (ms)', fontsize=18)
axes[1].tick_params(labelsize=15)
axes[1].legend(fontsize=13)
axes[1].grid(True)

# Plot third pair
axes[2].plot(list(range(len(lines[4]))), lines[4], marker='o', linestyle='-', color='b', label=f'Bloom: [{int(min(lines[4]))}, {int(max(lines[4]))}]')
axes[2].plot(list(range(len(lines[5]))), lines[5], marker='x', linestyle='-', color='r', label=f'Base: [{int(min(lines[5]))}, {int(max(lines[5]))}]')
# axes[2].set_title('Page Fetch Time (db: 1GB, page: 100)', fontsize=18)
axes[2].set_xlabel('Page Seq #', fontsize=18)
axes[2].xaxis.set_major_locator(MaxNLocator(integer=True))
axes[2].set_ylabel('Time (ms)', fontsize=18)
axes[2].tick_params(labelsize=15)
axes[2].legend(fontsize=13)
axes[2].grid(True)

# Adjust layout
plt.tight_layout()
plt.savefig("bloom_vary_pg.png", dpi=300, bbox_inches = "tight")
plt.show()
```
